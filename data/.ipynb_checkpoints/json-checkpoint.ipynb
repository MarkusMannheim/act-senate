```python
import pandas as pd, json, numpy as np

data = pd.read_csv(
    "2022_results.csv",
    usecols=[i for i in range(2, 10)],
    index_col=["candidateName"]
)
data_maddy = data.iloc[1, -3:]
data = data.groupby("PartyAb").sum().sort_values("first preferences", ascending=False)

sankey_data = {
    "nodes": [],
    "links": []
}
sankey_nocock = {
    "nodes": [],
    "links": []
}
for i, party in enumerate(data.index):
    sankey_data["nodes"].append({
        "node": int(i),
        "name": party,
        "party": party
    })
for i, preference in enumerate(data.columns[1:]):
    sankey_data["nodes"].append({
        "node": int(i + len(data)),
        "name": preference,
        "party": (
            "ALP" if i < 2 else
            "LP" if i == 2 else
            "DAVI" if i == 3 else "exhausted"
        )
    })
for i, party in enumerate(data.index):
    if i == 0:
        sankey_data["links"].append({
            "source": 0,
            "target": int(len(data)),
            "value": data.iat[0, 2]
        })
        # for j in range(3):
        #     maddy_data[j] > 0:
        #         sankey_data["links"].append({
        #             "source": 0,
        #             "target": int(len(data) + j + 1)
        #         })            
    else:
        for j in range(5):
            if data.iat[i, j + 1] > 0:
                sankey_data["links"].append({
                    "source": int(i),
                    "target": int(j + len(data)),
                    "value": data.iat[i, j + 1]
                })
# for i in range(3):
#     if data.iat[0, i + 3] > 0:
#         sankey_data["links"].append({
#             "source": int(len(data)),
#             "target": int(len(data) + i + 1),
#             "value": data_nocock.iat[0, i + 3]
#         })
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)
with open("./sankey_data_2022.json", "w") as file:
    file.write(json.dumps(sankey_data, cls=NpEncoder))
sankey_data
```
