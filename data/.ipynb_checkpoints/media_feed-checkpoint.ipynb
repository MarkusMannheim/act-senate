```python
from zipfile import ZipFile
import ftplib, os
import pandas as pd

# start FTP session
print("accessing FTP ...")
ftp = ftplib.FTP("mediafeedarchive.aec.gov.au")
ftp.login()
print(ftp.getwelcome())

# change working directory
ftp.cwd("24310/Standard/Light")

# access latest zip file
print("accessing latest results ...")
file_name = ftp.nlst()[-1]
directory_name = file_name.split('-')[-1][:-4]
try:
    os.mkdir(f"./feed_files/{directory_name}")
except:
    print("directory already exists")
local_file = open(f"./feed_files/{directory_name}/{file_name}", "wb")
print("saving to local file ...")
ftp.retrbinary("RETR " + file_name, local_file.write, 1024)
ftp.quit()
local_file.close()

# unzip latest file
zip_file = ZipFile(f"./feed_files/{directory_name}/{file_name}")
xml_file = zip_file.namelist()[-1]
zip_file.extract(xml_file, path=f"./feed_files/{directory_name}")

# read xml file
results = pd.read_xml(f"./feed_files/{directory_name}/{xml_file}", xpath=".//*")
for col in results.columns:
    print(col)
    print(results[col].describe())
```
