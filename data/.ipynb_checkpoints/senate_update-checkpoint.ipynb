```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt, matplotlib.ticker as ticker
sns.set(font_scale=1.25, style="whitegrid")

data = pd.read_csv("./aec-senate-formalpreferences-24310-ACT.csv")
parties = data.columns[6:]
for i in data.index:
    data.at[i, "voter_id"] = str(data.at[i, " Vote Collection Point ID"]).zfill(3) + str(data.at[i, " Batch No"]).zfill(3) + str(data.at[i, " Paper No"]).zfill(2)
data = data[data.columns[6:]]
data = data[[data.columns[-1]] + data.columns[:-1].tolist()]
for i in data.index:
    data.at[i, "no_votes"] = len(data.loc[i][1:].sort_values().dropna())
data[["prefs"]].to_csv("voters.csv")
```
