```python
import pandas as pd, numpy as np

choices = pd.read_csv(
    "2022_choices.csv",
    index_col="code"
)
choices["omitted"] = 0
choices["last"] = 0

votes = pd.read_csv(
    "2022_votes.csv",
    usecols=[1, 2],
    index_col=["voter id"]
)
votes = votes.sample(frac=1)

def vote_type(prefs):
    prefs = list(prefs)
    types = [choices.at[x, "btl"] for x in prefs]
    if (True in types) & (False in types):
        return np.nan
    elif True in types:
        return "b"
    else:
        return "a"

def above_check(prefs):
    for code in [x for x in choices.index if not choices.at[x, "btl"]]:
        if code not in prefs:
            choices.at[code, "omitted"] = choices.at[code, "omitted"] + 1
    if len(prefs) == 11:
        choices.at[prefs[-1], "last"] = choices.at[prefs[-1], "last"] + 1

def below_check(prefs):
    for code in [x for x in choices.index if choices.at[x, "btl"]]:
        if code not in prefs:
            choices.at[code, "omitted"] = choices.at[code, "omitted"] + 1
    if len(prefs) == 23:
        choices.at[prefs[-1], "last"] = choices.at[prefs[-1], "last"] + 1

for i in votes.index:
    prefs = votes.at[i, "prefs"]
    type_vote = vote_type(prefs)
    votes.at[i, "type"] = type_vote
    if type_vote == "a":
        above_check(prefs)
    if type_vote == "b":
        below_check(prefs)
```

```python
last = choices[choices["btl"]].sort_values("last", ascending=False)
last["candidateName"] = last["candidateName"].apply(lambda x: x.split(" ")[1] + " " + x.split(" ")[0])
last = last[["candidateName", "PartyAb", "last"]]
last.columns = ["candidate", "party", "last"]
last
```
