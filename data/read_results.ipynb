```python
import pandas as pd
import requests, PyPDF2

url = "https://results.aec.gov.au/24310/Website/External/SenateStateDop-24310-ACT.pdf"
response = requests.get(url)
raw_data = response.content

with open("latest_results.pdf", "wb") as data:
    data.write(raw_data)

pdf_file = open("latest_results.pdf", "rb")
pdf = PyPDF2.PdfFileReader(pdf_file)

text = []
for page in range(pdf.getNumPages()):
    if pdf.isEncrypted:
        pdf.decrypt("")
        print(pdf.getPage(page).extractText())
        page_text = pdf.getPage(page).extractText().split()
        text.append(page_text)
    else:
        print(pdf.getPage(page).extractText())
print(text)

# number_of_pages = reader.numPages
# page = reader.pages[0]
# text = page.extractText()
```
